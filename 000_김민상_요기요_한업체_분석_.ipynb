```python
import re
# 한글 텍스트 만 뽑아내기
# 텍스트 정제 함수 : 한글 이외의 문자는 전부 제거
def text_cleaning(text):
    # 한글의 정규표현식으로 한글만 추출합니다.
    hangul = re.compile('[^ ㄱ-ㅣ가-힣]+')   
    result = hangul.sub('', text)
    #지우기
    return result

#새로운 ko_text라는 것은 댓글을 클리닝에다 집어넣은 것이다. 
df2['ko_text'] = df2['댓글'].apply(lambda x: text_cleaning(x))

#원래 댓글이라는 변수 칸은 지워버린다
del df2['댓글']


# 그 중에서 한 글자 이상의 텍스트를 가지고 있는 데이터만 추출합니다
df2 = df2[df2['ko_text'].str.len() > 0]
df2.head()
```
